ESERCIZI:
1) scrivi un programma che calcoli il costo del biglietto di un traghetto poerando una distinzione tra autovetture e camion, ripartendo ulteriormente i veicoli per cilindrata:

AUTOVETTURA:
fino a 1000 cc. 20 euro
fino a 2000 cc. 30 euro
oltre 40 euro

CAMION:
fino a 2000 cc. 40 euro
fino a 3000 cc. 50 euro
oltre 100 euro

```python
tipo = str(input("tipo di veicolo (autovettura o camion)"))
while tipo != "autovettura" and tipo != "camion":
  tipo = str(input("tipo di veicolo (autovettura o camion)"))
cc = int(input("cilindrata del veicolo"))
while cc <= 0:
  cc = int(input("cilindrata del veicolo"))
prezzo = 0

if tipo == "autovettura":
  if cc <= 1000:
    prezzo = 20
  elif cc <= 2000:
    prezzo = 30
  else:
    prezzo = 40

elif tipo == "camion":
  if cc <= 2000:
    prezzo = 40
  elif cc <= 3000:
    prezzo = 50
  else:
    prezzo = 100

print(f"tipo: {tipo} cc: {cc} prezzo: {prezzo}")
```

Scrivi un programma che simuli il controllo di accesso a un sistema informatico.
Il sistema deve prima chiedere all'utente se è un amministratore o un utente normale.
Se l'utente è un amministratore, deve inserire una password di amministratore.
Se la password è corretta, il sistema stampa "Accesso amministratore consentito".
Se l'utente è un utente normale, deve inserire una password utente.
Se la password utente è corretta, il sistema stampa "Accesso utente consentito".
In caso di password errata, il sistema stampa "Accesso negato".
Assumi che la password di amministratore sia "admin123" e quella dell'utente sia "user123".

```python
admin_password = "admin123"
user_password = "user123"

ruolo = input("Sei un amministratore o un utente normale? (admin/utente): ").lower()

if ruolo == "admin":
    if input("Inserisci la password di amministratore: ") == admin_password:
        print("Accesso amministratore consentito")
    else:
        print("Accesso negato")
elif ruolo == "utente":
    if input("Inserisci la password utente: ") == user_password:
        print("Accesso utente consentito")
    else:
        print("Accesso negato")
else:
    print("Accesso negato")
```
