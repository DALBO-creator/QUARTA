# Esercizi Python
Questo notebook contiene:
1. Successione di Fibonacci
2. Somma dei numeri pari fino a N
3. Numeri pari inferiori a N (max 30) in ordine decrescente
4. Albero di Natale con carattere e altezza


```python
# ESERCIZIO 1: Successione di Fibonacci

num = int(input("Quanti termini della successione di Fibonacci vuoi visualizzare? "))

a, b = 1, 1  # primi due valori

print("Successione di Fibonacci:")
for i in range(num):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# ESERCIZIO 2: Somma dei numeri pari fino a num

num = int(input("Inserisci un numero: "))

somma = 0

for i in range(1, num + 1):
    if i % 2 == 0:
        somma += i

print("La somma dei numeri pari fino a", num, "Ã¨:", somma)
```

```python
# ESERCIZIO 3: Numeri pari inferiori a num, max 30, in ordine decrescente

num = int(input("Inserisci un numero: "))

print("Numeri pari positivi minori di", num, "in ordine decrescente (max 30):")

count = 0

for i in range(num - 1, 0, -1):
    if i % 2 == 0:
        print(i, end=" ")
        count += 1
        if count == 30:
            break
```

```python
# ESERCIZIO 4: Albero di Natale

c = input("Inserisci il carattere da usare per l'albero: ")
n = int(input("Inserisci l'altezza dell'albero: "))

for i in range(1, n + 1):
    spazi = " " * (n - i)
    simboli = c * (2 * i - 1)
    print(spazi + simboli)
```
