# Esercizi sulle strutture dati in Python
In questo notebook affronteremo:
1. Tupla con elementi di tipi diversi
2. Set: aggiunta, rimozione e controllo
3. Lista: aggiunta, inserimento e rimozione
4. Iterazione su un dizionario
5. Lista di dizionari (schedario studenti)
6. Dizionario con valori tuple (calendario esami)


```python
# Creazione tupla con 5 elementi di tipi diversi
mia_tupla = (42, "ciao", True, 3.14, [1, 2, 3])

# Stampa del terzo elemento (indice 2)
print("Terzo elemento:", mia_tupla[2])

# Controllo se la stringa "ciao" è presente
presente = "ciao" in mia_tupla
print("La stringa 'ciao' è presente?", presente)
```

```python
# Creazione del set
numeri = {1, 2, 3, 4, 5}

# Aggiunta del numero 6
numeri.add(6)

# Rimozione del numero 3
numeri.remove(3)

# Verifica se il numero 4 è presente
presente = 4 in numeri
print("Il numero 4 è presente?", presente)

# Stampa del set finale
print("Set finale:", numeri)
```

```python
# Creazione lista
frutti = ["mela", "banana", "arancia", "uva"]

# Aggiunta alla fine
frutti.append("kiwi")

# Inserimento alla seconda posizione (indice 1)
frutti.insert(1, "fragola")

# Rimozione di "banana"
frutti.remove("banana")

# Stampa lista finale
print("Lista finale:", frutti)
```

```python
# Dizionario prodotti
prodotti = {"latte": 1.5, "pane": 2.0, "uova": 3.0, "mela": 0.5}

# Iterazione e stampa
for prodotto, prezzo in prodotti.items():
    print(f"Il prezzo di {prodotto} è {prezzo}")
```

```python
# Dizionario iniziale
calendario_esami = {
    "Matematica": ("15/06/2024", "Aula 101"),
    "Fisica": ("18/06/2024", "Aula 102")
}

# Aggiunta di un nuovo esame
calendario_esami["Informatica"] = ("20/06/2024", "Aula 103")

# Modifica dell'aula di un esame (es. Fisica)
calendario_esami["Fisica"] = (calendario_esami["Fisica"][0], "Aula 202")

# Stampa di tutti gli esami
for materia, (data, aula) in calendario_esami.items():
    print(f"Esame di {materia}: Data = {data}, Aula = {aula}")
```
