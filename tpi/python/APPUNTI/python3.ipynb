# Tipi di dati in Python
Questo notebook mostra esempi di:
- array
- list
- tuple
- dict
- set


```python
from array import array

print("=== ARRAY (valori omogenei) ===")
numeri_interi = array("i", [5, 6, 3, 4])
print("Array di interi:", numeri_interi)

lettere = array("u", ["h", "a", "c"])
print("Array di lettere:", lettere)
```

```python
print("=== LIST (eterogenee e modificabili) ===")
lista_mista = [1, "ciao", True]
print(lista_mista)

# Modifica
lista_mista.append(3.14)
print("Dopo append:", lista_mista)
```

```python
print("=== TUPLE (non modificabili) ===")
giorni = ("lun", "mar", "mer", "gio", "ven", "sab", "dom")
print(giorni)

# Accesso
print("Primo giorno:", giorni[0])
```

```python
print("=== DICT (coppie chiave-valore) ===")
persona = {"nome": "Luca", "età": 30}
print(persona)

# Accesso
print("Nome:", persona["nome"])

# Modifica
persona["età"] = 31
print("Dopo modifica:", persona)
```

```python
print("=== SET (no duplicati) ===")
numeri = {1, 2, 3}
print(numeri)

numeri.add(4)
numeri.add(2)  # duplicato → ignorato
print("Dopo aggiunte:", numeri)

print("Il numero 3 è presente?", 3 in numeri)
```

# Riepilogo
- **Array**: ordinati, modificabili, omogenei
- **List**: ordinate, modificabili, eterogenee
- **Tuple**: ordinate, immutabili
- **Dict**: chiave → valore, modificabili
- **Set**: non ordinati, no duplicati
